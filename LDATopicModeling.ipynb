```python
# ==============================
# Step 1: Install dependencies
# ==============================
!pip install gensim pyLDAvis nltk "numpy<1.27" "scipy<1.14" --force-reinstall

# ==============================
# Step 2: Import libraries
# ==============================
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim import corpora
from gensim.models import LdaModel
import pyLDAvis.gensim_models
import pyLDAvis

# ==============================
# Step 3: Sample documents
# (You can replace this with your own dataset later)
# ==============================
```

```python
documents = [

    "The government announced new healthcare reforms to improve hospital facilities.",
    "Artificial intelligence and machine learning are advancing rapidly in the tech world.",
    "The stock market rose significantly after positive economic news from the finance ministry.",
    "Scientists discovered a new species of bird in the Amazon rainforest.",
    "Electric vehicles are gaining popularity as countries push for cleaner energy solutions.",
    "Global warming and climate change remain major concerns for world leaders.",
    "The new smartphone launch has attracted attention with its innovative features.",
    "Doctors recommend regular exercise and a balanced diet for better heart health.",
    "The football championship drew millions of viewers worldwide.",
    "A breakthrough in cancer research could lead to better treatment options.",
    "Cybersecurity threats are increasing as more companies move data online.",
    "The education ministry introduced new policies to improve digital learning in schools.",
    "Researchers are exploring deep learning techniques to improve natural language processing.",
    "Farmers are adopting smart agriculture technologies to increase crop yields.",
    "The airline industry is recovering after a sharp decline during the pandemic.",
    "Space agencies are planning new missions to explore Mars and beyond.",
    "E-commerce platforms reported record sales during the festive season.",
    "Music festivals are returning after restrictions were lifted in many countries.",
    "The United Nations held a summit to discuss sustainable development goals.",
    "Startups are driving innovation in renewable energy and green technology."

]

# ==============================
# Step 4: Preprocessing
# ==============================
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return tokens

processed_docs = [preprocess(doc) for doc in documents]
print("✅ Sample preprocessed doc:", processed_docs[0])

# ==============================
# Step 5: Create dictionary & corpus
# ==============================
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# ==============================
# Step 6: Build LDA model
# ==============================
lda_model = LdaModel(corpus=corpus,
                     id2word=dictionary,
                     num_topics=3,
                     random_state=42,
                     passes=10,
                     per_word_topics=True)

# Show discovered topics
print("\n🎯 LDA Topics:")
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")

# ==============================
# Step 7: Visualization (Inline + HTML)
# ==============================
from IPython.display import IFrame

# Prepare visualization
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)

# Show inline in Colab
pyLDAvis.display(vis)

# Save HTML for submission/sharing
pyLDAvis.save_html(vis, 'lda_topics.html')
print("\n✅ lda_topics.html file saved!")

# Show inside Colab using IFrame
display(IFrame('lda_topics.html', width=1000, height=600))

# Allow download
from google.colab import files
files.download('lda_topics.html')
```
